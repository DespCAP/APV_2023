### Functions

```python
import urllib.request
from PIL import Image
import matplotlib.pyplot as plt

# Function to print an image from a link:

def image_from_link(link,name):
  urllib.request.urlretrieve(link,name)
  img = Image.open(name)
  plt.figure()
  plt.imshow(img)
  plt.title(name)

# Usage
link = 'https://upload.wikimedia.org/wikipedia/commons/7/75/Trifid_Nebula_by_Deddy_Dayag.jpg'
name = 'Trifid Nebula'

image_from_link(link,name)
```

```python
# Display two graphs in a single row with 3 columns

import matplotlib.pyplot as plt

# a & a1 are displayed in the same graph
x = [0,1,2,3,4]
a = [0,1,2,1,0]
a1 = [0,1,1,1,0]

# p & p1 are displayed in the same graph
x1 = [0,0.001,1,1.5]
p = [0,2,2,1.5]
p1 = [0,1,1,1.5]

# First graph
plt.subplot(1, 3, 1)
plt.plot(x,a)
plt.plot(x,a1)

# Second graph
plt.subplot(1, 3, 2)
plt.plot(x1,p)
plt.plot(x1,p1)
```

```python
# A "for" loop can also be used for plotting
import matplotlib.pyplot as plt

x = [0,1,2,3,4]
y1 = [0,1,2,3,4]
y2 = [1,1,0,1,1]
y3 = [4,3,2,1,0]

#2d array:
array = [y1,y2,y3]
i=1

for element in array:
  plt.subplot(1, 3, i)
  plt.plot(x,element)
  i=i+1
```
